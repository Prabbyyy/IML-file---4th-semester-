```python
#question 3 

#5 coordinates declared
x=np.array([1,2,3,4,5]) 
y=np.array([2,4,5,4,5])

m,c=np.polyfit(x,y,1) #linear regression
y_ideal = m*x+c

total_error = np.sum((y-y_ideal)**2)

print("slope = ",m)
print("Intercept = ",c)
print("error = ", total_error)

```

```python
#question 3, derived 
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])   # x coordinates
y = np.array([2, 4, 5, 4, 5])   # y coordinates

# Calculations
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / n

y_ideal = m * x + c
total_error = np.sum((y - y_ideal) ** 2)

# Output
print(f"Slope = {m:.2f}")
print(f"Intercept = {c:.2f}")
print(f"Error = {total_error:.2f}")

# Plotting
plt.scatter(x, y)                 # actual data points
plt.plot(x, y_ideal)              # regression line
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Graph:")
plt.grid(True)
plt.show()
```
