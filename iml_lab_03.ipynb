```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


# load dataset
data = pd.read_csv('daily-minimum-temperatures-in-me (1).csv')


# convert columns
data["Date"] = pd.to_datetime(data["Date"], errors="coerce")

data["Daily minimum temperatures"] = pd.to_numeric(
    data["Daily minimum temperatures"], errors="coerce"
)


# remove missing values
data = data.dropna()


# create numeric index for days
data["day_index"] = (data["Date"] - data["Date"].min()).dt.days


# features and target
X = data[["day_index"]]
y = data["Daily minimum temperatures"]


# train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


# model using split data
model_split = LinearRegression()
model_split.fit(X_train, y_train)

y_pred_split = model_split.predict(X_test)

r2_split = r2_score(y_test, y_pred_split)

print("R2 Score (Train-Test Split) without rounding:", r2_split)
print("R2 Score (Train-Test Split) rounded:", round(r2_split, 2))


# model using full dataset
model_full = LinearRegression()

model_full.fit(X, y)

y_pred_full = model_full.predict(X)

r2_full = r2_score(y, y_pred_full)

print("R2 Score (Full Data) without rounding:", r2_full)
print("R2 Score (Full Data) rounded:", round(r2_full, 2))
```

```python
#question-02
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

class MeraLR:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):
        X_train = np.array(X_train).flatten()
        y_train = np.array(y_train)

        xm = np.mean(X_train)
        ym = np.mean(y_train)

        numerator = np.sum((X_train - xm) * (y_train - ym))
        denominator = np.sum((X_train - xm) ** 2)

        self.m = numerator / denominator
        self.b = ym - self.m * xm

    def predict(self, X_test):
        X_test = np.array(X_test).flatten()
        y_pred = self.m * X_test + self.b
        return y_pred

    def r2_score(self, y_true, y_pred):
        y_true = np.array(y_true)

        total_ss = np.sum((y_true - np.mean(y_true)) ** 2)
        residual_ss = np.sum((y_true - y_pred) ** 2)

        r2 = 1 - (residual_ss / total_ss)
        return r2


# reading dataset
data = pd.read_csv("daily-minimum-temperatures-in-me (1).csv")

# cleaning
data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
data["Daily minimum temperatures"] = pd.to_numeric(
    data["Daily minimum temperatures"], errors="coerce"
)

data = data.dropna()

# feature creation
data["day_index"] = (data["Date"] - data["Date"].min()).dt.days

X = data[["day_index"]]
y = data["Daily minimum temperatures"]

# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# train
model = MeraLR()
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# evaluate
r2 = model.r2_score(y_test, y_pred)

print("Slope (m) without rounding =", model.m)
print("Slope (m) rounded =", round(model.m, 2))

print("Intercept (b) without rounding =", model.b)
print("Intercept (b) rounded =", round(model.b, 2))

print("R2 Score without rounding =", r2)
print("R2 Score rounded =", round(r2, 2))
```

```python
#question-03

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

data = pd.read_csv("daily-minimum-temperatures-in-me (2).csv")

data.columns = data.columns.str.strip()

data = data.drop(columns=["Unnamed: 2", "Unnamed: 3"], errors="ignore")

data["Date"] = pd.to_datetime(data["Date"], errors="coerce")

data["Daily minimum temperatures"] = pd.to_numeric(
    data["Daily minimum temperatures"], errors="coerce"
)

data["Humidity"] = pd.to_numeric(data["Humidity"], errors="coerce")

data = data.dropna()

data["day_index"] = (data["Date"] - data["Date"].min()).dt.days

X = data[["day_index", "Humidity"]]
y = data["Daily minimum temperatures"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model_1 = LinearRegression()

model_1.fit(X_train, y_train)

y_pred1 = model_1.predict(X_test)

r2_split = r2_score(y_test, y_pred1)

print("R2 Score (Train-Test Split) without rounding =", r2_split)
print("R2 Score (Train-Test Split) rounded =", round(r2_split, 2))


model_2 = LinearRegression()

model_2.fit(X, y)

y_pred2 = model_2.predict(X)

r2_full = r2_score(y, y_pred2)

print("R2 Score (Full Data) without rounding =", r2_full)
print("R2 Score (Full Data) rounded =", round(r2_full, 2))
```

```python
#question 4 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

class MeraMultiLR:
    def __init__(self):
        self.w0 = None
        self.w1 = None
        self.w2 = None

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        ones = np.ones((X.shape[0], 1))
        X_b = np.hstack((ones, X))

        W = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

        self.w0 = W[0]
        self.w1 = W[1]
        self.w2 = W[2]

    def predict(self, X):
        X = np.array(X)
        return self.w0 + self.w1 * X[:, 0] + self.w2 * X[:, 1]

    def r2_score(self, y_true, y_pred):
        y_true = np.array(y_true)

        ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
        ss_residual = np.sum((y_true - y_pred) ** 2)

        return 1 - (ss_residual / ss_total)


# load dataset
data = pd.read_csv("daily-minimum-temperatures-in-me (2).csv")

# clean columns
data.columns = data.columns.str.strip()
data = data.drop(columns=["Unnamed: 2", "Unnamed: 3"], errors="ignore")

data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
data["Daily minimum temperatures"] = pd.to_numeric(
    data["Daily minimum temperatures"], errors="coerce"
)
data["Humidity"] = pd.to_numeric(data["Humidity"], errors="coerce")

data = data.dropna()

# feature engineering
data["day_index"] = (data["Date"] - data["Date"].min()).dt.days

X = data[["day_index", "Humidity"]]
y = data["Daily minimum temperatures"]

# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# train model
model = MeraMultiLR()
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# evaluate
r2 = model.r2_score(y_test, y_pred)

# outputs
print("w0 (Intercept) without rounding =", model.w0)
print("w0 (Intercept) rounded =", round(model.w0, 2))

print("w1 (day_index) without rounding =", model.w1)
print("w1 (day_index) rounded =", round(model.w1, 2))

print("w2 (Humidity) without rounding =", model.w2)
print("w2 (Humidity) rounded =", round(model.w2, 2))

print("R2 Score without rounding =", r2)
print("R2 Score rounded =", round(r2, 2))
```
